```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from skimage.transform import resize
from IPython.display import HTML
import warnings
warnings.filterwarnings("ignore")
from google.colab import drive


drive.mount('/content/drive')

source_image = imageio.imread("/content/drive/MyDrive/test/warren.jpg")
driver_video = imageio.get_reader('/content/drive/MyDrive/test/abarnvbtwb.mp4')


source_image = resize(source_image, (256, 256))[..., :3]
driver_video = [resize(frame, (256, 256))[..., :3] for frame in driver_video] #resize each image/frame in the video

def display(source, driving, generated=None):
    fig = plt.figure(figsize=(8 + 4 * (generated is not None), 6))

    arr = []
    for i in range(len(driving)):
        cols = [source] #source=still image
        cols.append(driving[i])#appending each frame/img to list 'cols'
        if generated is not None:
            cols.append(generated[i]) #if generated img/frame(from generated video) is not none then append
        visual = plt.imshow(np.concatenate(cols, axis=1), animated=True) #
        plt.axis('off')
        arr.append([visual])

    anime = animation.ArtistAnimation(fig, arr, interval=50, repeat_delay=1000)
    plt.close()
    return anime
```

```python
HTML(display(source_image, driver_video).to_html5_video())
```
